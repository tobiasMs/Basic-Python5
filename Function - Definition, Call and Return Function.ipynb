```python
def devide_by_2(x):
  devide = x/2
  return devide
```

###Pass by Refrence vs Value

*   All argument in python is "passed by reference". that's mean when we change a variable, the referencing data will changed in inside function or outside function.
*   Except if we use operator assignment will changed reference parameter.



```python
def change(a_list):
  print("Value of the function is {}".format(a_list))
```

```python
a_list = [1,2]
```

```python
change(a_list)
```

```python
a_list.append([3,4])
change(a_list)
```

list in the function (inside def function) called LOCAL, and the append one (outside def function) called GLOBAL

```python
def change(a_list):
  a_list = [1,2]
  print("The Local value of the function is {}".format(a_list))
```

```python
a_list = [3,4]
change(a_list)
print("The Global value of the function is {}".format(a_list))
```
